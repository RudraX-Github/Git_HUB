```m
// Step 1: Load and change data types
let
    Source = Csv.Document(File.Contents("ML_Predictions.csv"), [Delimiter=",", Columns=8, Encoding=1252, QuoteStyle=QuoteStyle.None]),
    #"Promoted Headers" = Table.PromoteHeaders(Source, [PromoteAllScalars=true]),
    
    // Step 2: Transform column types
    #"Changed Type" = Table.TransformColumnTypes(
        #"Promoted Headers",
        {
            {"Date", type date},
            {"Commodity", type text},
            {"Market", type text},
            {"Actual_Price", type number},
            {"Predicted_Price", type number},
            {"Prediction_Error", type number},
            {"Error_Percentage", type number},
            {"Model_Name", type text}
        }
    ),
    
    // Step 3: Add Absolute Error Column
    #"Added Abs_Error" = Table.AddColumn(
        #"Changed Type",
        "Absolute_Error",
        each Number.Abs([Prediction_Error]),
        type number
    ),
    
    // Step 4: Add Accuracy Category
    #"Added Accuracy_Category" = Table.AddColumn(
        #"Added Abs_Error",
        "Accuracy_Category",
        each if [Error_Percentage] <= 5 then "Excellent (<5%)"
             else if [Error_Percentage] <= 10 then "Good (5-10%)"
             else if [Error_Percentage] <= 15 then "Fair (10-15%)"
             else "Poor (>15%)",
        type text
    ),
    
    // Step 5: Add Prediction Direction (Over/Under prediction)
    #"Added Direction" = Table.AddColumn(
        #"Added Accuracy_Category",
        "Prediction_Direction",
        each if [Prediction_Error] > 0 then "Over-Predicted"
             else if [Prediction_Error] < 0 then "Under-Predicted"
             else "Exact",
        type text
    ),
    
    // Step 6: Add Month and Quarter for analysis
    #"Added Month" = Table.AddColumn(
        #"Added Direction",
        "Month",
        each Date.Month([Date]),
        Int64.Type
    ),
    
    #"Added Quarter" = Table.AddColumn(
        #"Added Month",
        "Quarter",
        each Date.QuarterOfYear([Date]),
        Int64.Type
    )
in
    #"Added Quarter"
```

```m
// Step 1: Load and set data types
let
    Source = Csv.Document(File.Contents("Model_Performance.csv"), [Delimiter=",", Columns=6]),
    #"Promoted Headers" = Table.PromoteHeaders(Source, [PromoteAllScalars=true]),
    
    // Step 2: Transform types
    #"Changed Type" = Table.TransformColumnTypes(
        #"Promoted Headers",
        {
            {"Model_Name", type text},
            {"MAE", type number},
            {"RMSE", type number},
            {"R2_Score", type number},
            {"MAPE", type number},
            {"Training_Time", type number}
        }
    ),
    
    // Step 3: Add Performance Rating
    #"Added Rating" = Table.AddColumn(
        #"Changed Type",
        "Performance_Rating",
        each if [R2_Score] >= 0.90 then "Excellent (>90%)"
             else if [R2_Score] >= 0.85 then "Very Good (85-90%)"
             else if [R2_Score] >= 0.70 then "Good (70-85%)"
             else "Needs Improvement (<70%)",
        type text
    ),
    
    // Step 4: Calculate Error Metrics Rank
    #"Added Rank" = Table.AddColumn(
        #"Added Rating",
        "Model_Rank",
        each Table.PositionOf(
            Table.Sort(#"Added Rating", {{"R2_Score", Order.Descending}}),
            _
        ) + 1,
        Int64.Type
    )
in
    #"Added Rank"
```

### **1.4 Create Calendar Table (DAX)**

```dax
Calendar = 
ADDCOLUMNS(
    CALENDAR(DATE(2024,1,1), DATE(2025,1,31)),
    "Year", YEAR([Date]),
    "Month", FORMAT([Date], "MMM"),
    "MonthNum", MONTH([Date]),
    "MonthYear", FORMAT([Date], "MMM YYYY"),
    "Quarter", "Q" & FORMAT([Date], "Q"),
    "QuarterNum", QUARTER([Date]),
    "WeekNum", WEEKNUM([Date]),
    "DayName", FORMAT([Date], "dddd"),
    "DayNum", DAY([Date]),
    "DayOfWeek", WEEKDAY([Date]),
    "IsWeekend", IF(WEEKDAY([Date]) IN {1,7}, "Weekend", "Weekday"),
    "IsHoliday", IF([Date] IN {DATE(2024,1,26), DATE(2024,8,15), DATE(2024,10,2)}, "Holiday", "Regular")
)
```

## ğŸ“Š **STEP 2: Create DAX Measures**

### **2.1 Core Price Metrics**

```dax
// Total Average Price
Avg_Price = AVERAGE(Price_Data_Processed[Modal_Price])

// Total Records
Total_Records = COUNTROWS(Price_Data_Processed)

// Min Price (Overall)
Min_Price_Overall = MIN(Price_Data_Processed[Min_Price])

// Max Price (Overall)
Max_Price_Overall = MAX(Price_Data_Processed[Max_Price])

// Average Volatility
Avg_Volatility = AVERAGE(Price_Data_Processed[Price_Volatility_%])

// Total Markets
Total_Markets = DISTINCTCOUNT(Price_Data_Processed[Market])

// Total Commodities
Total_Commodities = DISTINCTCOUNT(Price_Data_Processed[Commodity])

// Price Range
Price_Range = [Max_Price_Overall] - [Min_Price_Overall]
```

---

### **2.2 Time Intelligence Measures**

```dax
// Previous Month Average Price
Prev_Month_Price = 
CALCULATE(
    [Avg_Price],
    DATEADD(Calendar[Date], -1, MONTH)
)

// Month-over-Month Change
MoM_Change = [Avg_Price] - [Prev_Month_Price]

// MoM % Change
MoM_Change_% = 
DIVIDE(
    [MoM_Change],
    [Prev_Month_Price],
    0
) * 100

// Year-to-Date Average
YTD_Avg_Price = 
CALCULATE(
    [Avg_Price],
    DATESYTD(Calendar[Date])
)

// Moving Average (7-Day)
Moving_Avg_7D = 
CALCULATE(
    [Avg_Price],
    DATESINPERIOD(
        Calendar[Date],
        LASTDATE(Calendar[Date]),
        -7,
        DAY
    )
)

// Moving Average (30-Day)
Moving_Avg_30D = 
CALCULATE(
    [Avg_Price],
    DATESINPERIOD(
        Calendar[Date],
        LASTDATE(Calendar[Date]),
        -30,
        DAY
    )
)
```

---

### **2.3 Weather-Related Measures**

```dax
// Average Temperature
Avg_Temperature = AVERAGE(Weather[temperature_2m_mean])

// Max Temperature
Max_Temperature = MAX(Weather[temperature_2m_max])

// Min Temperature
Min_Temperature = MIN(Weather[temperature_2m_min])

// Total Rainfall
Total_Rainfall = SUM(Weather[rain_sum])

// Rainy Days Count
Rainy_Days = 
CALCULATE(
    COUNTROWS(Weather),
    Weather[Is_Rainy] = "Yes"
)

// Average Wind Speed
Avg_Wind_Speed = AVERAGE(Weather[wind_speed_10m_max])

// Clear Days %
Clear_Days_% = 
DIVIDE(
    CALCULATE(
        COUNTROWS(Weather),
        Weather[Weather_Description] = "Clear Sky"
    ),
    COUNTROWS(Weather),
    0
) * 100
```

---

### **2.4 Price Performance Indicators**

```dax
// High Price Count (Above 5000)
High_Price_Count = 
CALCULATE(
    COUNTROWS(Price_Data_Processed),
    Price_Data_Processed[Modal_Price] > 5000
)

// High Price %
High_Price_% = 
DIVIDE(
    [High_Price_Count],
    [Total_Records],
    0
) * 100

// Price Trend (Rising/Falling)
Price_Trend = 
IF(
    [MoM_Change] > 0,
    "Rising â†‘",
    IF(
        [MoM_Change] < 0,
        "Falling â†“",
        "Stable â†’"
    )
)

// Price Stability Score (Lower volatility = Higher stability)
Price_Stability_Score = 
100 - MIN([Avg_Volatility], 100)

// Top Commodity by Price
Top_Commodity = 
CALCULATE(
    FIRSTNONBLANK(Price_Data_Processed[Commodity], 1),
    TOPN(
        1,
        ALL(Price_Data_Processed[Commodity]),
        [Avg_Price],
        DESC
    )
)

// Top Market by Records
Top_Market = 
CALCULATE(
    FIRSTNONBLANK(Price_Data_Processed[Market], 1),
    TOPN(
        1,
        ALL(Price_Data_Processed[Market]),
        [Total_Records],
        DESC
    )
)
```

---

### **2.5 Weather-Price Correlation Indicators**

```dax
// High Temp Days with High Prices
High_Temp_High_Price = 
CALCULATE(
    [Avg_Price],
    Weather[temperature_2m_max] > 35
)

// Rainy Days Average Price
Rainy_Day_Avg_Price = 
CALCULATE(
    [Avg_Price],
    Weather[Is_Rainy] = "Yes"
)

// Non-Rainy Days Average Price
Non_Rainy_Day_Avg_Price = 
CALCULATE(
    [Avg_Price],
    Weather[Is_Rainy] = "No"
)

// Rain Price Impact
Rain_Price_Impact = [Rainy_Day_Avg_Price] - [Non_Rainy_Day_Avg_Price]

// Rain Impact %
Rain_Impact_% = 
DIVIDE(
    [Rain_Price_Impact],
    [Non_Rainy_Day_Avg_Price],
    0
) * 100
```

---

### **2.6 Market Performance Metrics**

```dax
// Market Share %
Market_Share_% = 
DIVIDE(
    CALCULATE([Total_Records]),
    CALCULATE([Total_Records], ALL(Price_Data_Processed[Market])),
    0
) * 100

// Commodity Diversity (Unique commodities per market)
Commodity_Diversity = 
CALCULATE(
    DISTINCTCOUNT(Price_Data_Processed[Commodity])
)

// Price Premium Over Average
Price_Premium = 
[Avg_Price] - 
CALCULATE(
    [Avg_Price],
    ALL(Price_Data_Processed)
)

// Premium %
Premium_% = 
DIVIDE(
    [Price_Premium],
    CALCULATE([Avg_Price], ALL(Price_Data_Processed)),
    0
) * 100
```

### **5.3 Anomaly Detection**

**Create Anomaly Detection Measure:**
```dax
Price_Anomaly = 
VAR AvgPrice = [Avg_Price]
VAR StdDev = STDEVX.P(ALL(Price_Data_Processed), Price_Data_Processed[Modal_Price])
VAR Threshold = 2 * StdDev
VAR GlobalAvg = CALCULATE([Avg_Price], ALL(Price_Data_Processed))

RETURN
    IF(
        ABS(AvgPrice - GlobalAvg) > Threshold,
        "ğŸš¨ Anomaly Detected",
        "âœ… Normal"
    )

// Z-Score Calculation
Price_ZScore = 
VAR CurrentPrice = [Avg_Price]
VAR MeanPrice = CALCULATE([Avg_Price], ALL(Price_Data_Processed))
VAR StdDev = STDEVX.P(ALL(Price_Data_Processed), Price_Data_Processed[Modal_Price])

RETURN
    DIVIDE(CurrentPrice - MeanPrice, StdDev, 0)

// Anomaly Flag (Z-Score > 2 or < -2)
Is_Anomaly = IF(ABS([Price_ZScore]) > 2, 1, 0)
```

### **6.8 Performance Optimization**

**Optimization Techniques:**

```
1. Reduce Visual Count:
   âŒ Don't: 20+ visuals per page
   âœ… Do: Maximum 10-12 visuals per page

2. Use Aggregated Data:
   - Import pre-aggregated tables for large datasets
   - Use summary tables (already have: Commodity_Summary, Market_Summary)

3. Limit Interactions:
   - Disable cross-filtering where not needed
   - Use "None" interaction for static visuals

4. Optimize DAX:
   - Use variables (VAR) to avoid repeated calculations
   - Use SELECTEDVALUE instead of VALUES with HASONEVALUE
   - Avoid row-by-row calculations

5. Background Refresh:
   - Set slicers to load on page open only
   - Use query reduction for large datasets
   - Enable "Apply" button for filters

6. Visual Level Filters:
   - Apply filters at visual level instead of page level when possible
   - Use Top N filters to limit data processed

7. Use Bookmarks for Scenarios:
   - Instead of showing all data, use bookmarks to switch views
   - Reduces memory load

8. Compress Images:
   - Background images: Max 1920x1080, compressed
   - Icons: SVG format preferred
   - Logos: PNG with transparency, <100KB
```

### **9.1 DAX Optimization Techniques**

**Use Variables:**
```dax
// âŒ BAD: Repeated calculations
Inefficient_Measure = 
    IF(
        CALCULATE(SUM(Price_Data_Processed[Modal_Price])) > 5000,
        CALCULATE(SUM(Price_Data_Processed[Modal_Price])) * 1.1,
        CALCULATE(SUM(Price_Data_Processed[Modal_Price]))
    )

// âœ… GOOD: Use variables
Efficient_Measure = 
VAR TotalPrice = CALCULATE(SUM(Price_Data_Processed[Modal_Price]))
RETURN
    IF(
        TotalPrice > 5000,
        TotalPrice * 1.1,
        TotalPrice
    )
```

**Avoid Row Context:**
```dax
// âŒ BAD: Row-by-row calculation (slow)
Slow_Measure = 
    SUMX(
        Price_Data_Processed,
        IF(
            Price_Data_Processed[Modal_Price] > 5000,
            Price_Data_Processed[Modal_Price],
            0
        )
    )

// âœ… GOOD: Filter context (fast)
Fast_Measure = 
    CALCULATE(
        SUM(Price_Data_Processed[Modal_Price]),
        Price_Data_Processed[Modal_Price] > 5000
    )
```

**Materialize Complex Calculations:**
```dax
// Instead of complex DAX, add calculated column in Power Query

Power Query M:
= Table.AddColumn(
    Source, 
    "Price_Category",
    each if [Modal_Price] >= 7000 then "Premium"
         else if [Modal_Price] >= 4000 then "High"
         else if [Modal_Price] >= 2000 then "Medium"
         else "Low"
)
```

```m
// âœ… GOOD: These operations fold back to source
= Source
  |> Table.SelectRows(each [Modal_Price] > 1000)  // Folds
  |> Table.SelectColumns({"Commodity", "Market", "Modal_Price"})  // Folds
  |> Table.Group(...)  // Folds (if simple)

// âŒ BAD: These prevent query folding
= Source
  |> Table.AddColumn(each [Custom_Calculation])  // Doesn't fold
  |> Table.Sort(...)  // After non-folding step
```
