## ğŸ“Š **STEP 2: Create DAX Measures (ML-Focused)**

### **2.1 Model Performance Metrics**

```dax
// ========================================
// KEY MODEL PERFORMANCE INDICATORS (KPIs)
// ========================================

// Best Model RÂ² Score
Best_R2_Score = 
MAXX(Model_Performance, Model_Performance[R2_Score])

// Best Model Name
Best_Model = 
CALCULATE(
    FIRSTNONBLANK(Model_Performance[Model_Name], 1),
    TOPN(1, ALL(Model_Performance), [R2_Score], DESC)
)

// Average RÂ² Across All Models
Avg_R2_Score = 
AVERAGE(Model_Performance[R2_Score])

// Best Model MAE (Mean Absolute Error)
Best_MAE = 
MINX(Model_Performance, Model_Performance[MAE])

// Best Model RMSE (Root Mean Squared Error)
Best_RMSE = 
MINX(Model_Performance, Model_Performance[RMSE])

// Best Model MAPE (Mean Absolute Percentage Error)
Best_MAPE = 
MINX(Model_Performance, Model_Performance[MAPE])

// Model Performance Grade (Based on RÂ²)
Model_Grade = 
SWITCH(
    TRUE(),
    [Best_R2_Score] >= 0.90, "A+ (Excellent)",
    [Best_R2_Score] >= 0.85, "A (Very Good)",
    [Best_R2_Score] >= 0.75, "B (Good)",
    [Best_R2_Score] >= 0.65, "C (Fair)",
    "D (Needs Improvement)"
)

// Total Training Time (All Models)
Total_Training_Time = 
SUM(Model_Performance[Training_Time])
```

### **2.2 Prediction Accuracy Metrics**

```dax
// ========================================
// PREDICTION ACCURACY & ERROR ANALYSIS
// ========================================

// Total Predictions Count
Total_Predictions = 
COUNTROWS(ML_Predictions)

// Average Actual Price (Test Set)
Avg_Actual_Price = 
AVERAGE(ML_Predictions[Actual_Price])

// Average Predicted Price
Avg_Predicted_Price = 
AVERAGE(ML_Predictions[Predicted_Price])

// Average Prediction Error (Signed)
Avg_Prediction_Error = 
AVERAGE(ML_Predictions[Prediction_Error])

// Average Absolute Error
Avg_Absolute_Error = 
AVERAGE(ML_Predictions[Absolute_Error])

// Mean Absolute Percentage Error (MAPE) - Calculated
Calculated_MAPE = 
AVERAGEX(
    ML_Predictions,
    ABS(ML_Predictions[Prediction_Error]) / ML_Predictions[Actual_Price]
) * 100

// Prediction Accuracy %
Prediction_Accuracy = 
(1 - ([Avg_Absolute_Error] / [Avg_Actual_Price])) * 100

// Max Prediction Error
Max_Error = 
MAX(ML_Predictions[Absolute_Error])

// Min Prediction Error
Min_Error = 
MIN(ML_Predictions[Absolute_Error])

// Prediction Error Std Dev
Error_StdDev = 
STDEV.P(ML_Predictions[Prediction_Error])

// Excellent Predictions Count (<5% error)
Excellent_Predictions = 
CALCULATE(
    [Total_Predictions],
    ML_Predictions[Accuracy_Category] = "Excellent (<5%)"
)

// Excellent Predictions %
Excellent_Predictions_% = 
DIVIDE([Excellent_Predictions], [Total_Predictions], 0) * 100

// Poor Predictions Count (>15% error)
Poor_Predictions = 
CALCULATE(
    [Total_Predictions],
    ML_Predictions[Accuracy_Category] = "Poor (>15%)"
)

// Poor Predictions %
Poor_Predictions_% = 
DIVIDE([Poor_Predictions], [Total_Predictions], 0) * 100

// Over-Prediction Count
Over_Predictions = 
CALCULATE(
    [Total_Predictions],
    ML_Predictions[Prediction_Direction] = "Over-Predicted"
)

// Under-Prediction Count
Under_Predictions = 
CALCULATE(
    [Total_Predictions],
    ML_Predictions[Prediction_Direction] = "Under-Predicted"
)

// Over/Under Prediction Ratio
Prediction_Bias = 
DIVIDE([Over_Predictions], [Under_Predictions], 1)
```

### **2.3 Commodity & Market Performance**

```dax
// ========================================
// COMMODITY-SPECIFIC ACCURACY METRICS
// ========================================

// Avg Error by Commodity (Use in table visual)
Avg_Error_by_Commodity = 
CALCULATE(
    [Avg_Absolute_Error],
    ALLEXCEPT(ML_Predictions, ML_Predictions[Commodity])
)

// Best Predicted Commodity (Lowest Error)
Best_Predicted_Commodity = 
CALCULATE(
    FIRSTNONBLANK(ML_Predictions[Commodity], 1),
    TOPN(
        1,
        SUMMARIZE(
            ML_Predictions,
            ML_Predictions[Commodity],
            "AvgError", [Avg_Absolute_Error]
        ),
        [AvgError],
        ASC
    )
)

// Worst Predicted Commodity (Highest Error)
Worst_Predicted_Commodity = 
CALCULATE(
    FIRSTNONBLANK(ML_Predictions[Commodity], 1),
    TOPN(
        1,
        SUMMARIZE(
            ML_Predictions,
            ML_Predictions[Commodity],
            "AvgError", [Avg_Absolute_Error]
        ),
        [AvgError],
        DESC
    )
)

// Avg Error by Market
Avg_Error_by_Market = 
CALCULATE(
    [Avg_Absolute_Error],
    ALLEXCEPT(ML_Predictions, ML_Predictions[Market])
)

// Market Prediction Accuracy Rank
Market_Accuracy_Rank = 
RANKX(
    ALL(ML_Predictions[Market]),
    [Avg_Absolute_Error],
    ,
    ASC,
    Dense
)
```

### **2.5 Weather-Price Correlation Metrics**

```dax
// ========================================
// WEATHER IMPACT ON PREDICTIONS
// ========================================

// Average Temperature (Test Period)
Avg_Temperature = 
AVERAGE(Weather[temperature_2m_mean])

// Max Temperature (Test Period)
Max_Temperature = 
MAX(Weather[temperature_2m_max])

// Min Temperature (Test Period)
Min_Temperature = 
MIN(Weather[temperature_2m_min])

// Total Rainfall (Test Period)
Total_Rainfall = 
SUM(Weather[rain_sum])

// Rainy Days Count
Rainy_Days_Count = 
CALCULATE(
    COUNTROWS(Weather),
    Weather[Is_Rainy] = "Yes"
)

// Rainy Days %
Rainy_Days_% = 
DIVIDE(
    [Rainy_Days_Count],
    COUNTROWS(Weather),
    0
) * 100

// Avg Price on Rainy Days
Avg_Price_Rainy = 
CALCULATE(
    [Avg_Predicted_Price],
    Weather[Is_Rainy] = "Yes"
)

// Avg Price on Non-Rainy Days
Avg_Price_NonRainy = 
CALCULATE(
    [Avg_Predicted_Price],
    Weather[Is_Rainy] = "No"
)

// Rain Impact on Price (Rs)
Rain_Impact = 
[Avg_Price_Rainy] - [Avg_Price_NonRainy]

// Rain Impact %
Rain_Impact_% = 
DIVIDE([Rain_Impact], [Avg_Price_NonRainy], 0) * 100

// Avg Error on High Temp Days (>35Â°C)
Avg_Error_HighTemp = 
CALCULATE(
    [Avg_Absolute_Error],
    Weather[temperature_2m_max] > 35
)

// Avg Error on Low Temp Days (<20Â°C)
Avg_Error_LowTemp = 
CALCULATE(
    [Avg_Absolute_Error],
    Weather[temperature_2m_min] < 20
)

// Temperature-Error Correlation Flag
Temp_Error_Correlation = 
IF(
    [Avg_Error_HighTemp] > [Avg_Absolute_Error] * 1.2,
    "High Temp = Higher Error",
    IF(
        [Avg_Error_LowTemp] > [Avg_Absolute_Error] * 1.2,
        "Low Temp = Higher Error",
        "No Strong Correlation"
    )
)
```
