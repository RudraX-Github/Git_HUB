### Step 3: Create Calendar Table

#### DAX Formula for Date Table

```DAX
Date_Table = 
ADDCOLUMNS (
    CALENDAR(DATE(2024, 1, 1), DATE(2025, 1, 1)),
    "Year", YEAR([Date]),
    "Month", MONTH([Date]),
    "Month Name", FORMAT([Date], "MMMM"),
    "Month Short", FORMAT([Date], "MMM"),
    "Quarter", "Q" & FORMAT([Date], "Q"),
    "Week Number", WEEKNUM([Date]),
    "Day Name", FORMAT([Date], "dddd"),
    "Day Short", FORMAT([Date], "ddd"),
    "Day of Year", [Date] - DATE(YEAR([Date]), 1, 1) + 1,
    "Is Weekend", IF(WEEKDAY([Date]) IN {1, 7}, TRUE, FALSE)
)
```

**Mark as Date Table**: Right-click ‚Üí Mark as Date Table

```M
= Table.AddColumn(
    PreviousStep, 
    "Price Category", 
    each if [Modal_Price] < 2000 then "Low"
         else if [Modal_Price] < 5000 then "Medium"
         else if [Modal_Price] < 10000 then "High"
         else "Premium"
)
```

## üßÆ 6. DAX Measures & Calculations <a id="dax-measures"></a>

### Essential Measures for Dashboard

Create these measures in Power BI by going to **Modeling ‚Üí New Measure**

---

#### Basic Aggregations

```DAX
// Total Records
Total Records = COUNT(Price_Data[Modal_Price])

// Average Price
Average Price = AVERAGE(Price_Data[Modal_Price])

// Total Commodities
Total Commodities = DISTINCTCOUNT(Price_Data[Commodity])

// Total Markets
Total Markets = DISTINCTCOUNT(Price_Data[Market])

// Total Trading Value (approximate)
Total Trading Value = SUMX(Price_Data, Price_Data[Modal_Price] * 1)
```

---

#### Price Analysis Measures

```DAX
// Minimum Price
Min Price = MIN(Price_Data[Min_Price])

// Maximum Price
Max Price = MAX(Price_Data[Max_Price])

// Price Range
Price Range = [Max Price] - [Min Price]

// Average Volatility
Average Volatility = AVERAGE(Price_Data[Price_Volatility_%])

// Weighted Average Price (by volume, if available)
Weighted Avg Price = 
DIVIDE(
    SUMX(Price_Data, Price_Data[Modal_Price] * 1),
    COUNT(Price_Data[Modal_Price])
)
```

---

#### Time Intelligence Measures

```DAX
// Previous Month Price
Previous Month Price = 
CALCULATE(
    [Average Price],
    DATEADD(Date_Table[Date], -1, MONTH)
)

// Month-over-Month Change
MoM Change = [Average Price] - [Previous Month Price]

// MoM % Change
MoM % Change = 
DIVIDE(
    [MoM Change],
    [Previous Month Price],
    0
)

// Year-to-Date Average
YTD Average Price = 
CALCULATE(
    [Average Price],
    DATESYTD(Date_Table[Date])
)

// Moving Average 7-Day
MA7 = 
CALCULATE(
    [Average Price],
    DATESINPERIOD(Date_Table[Date], LASTDATE(Date_Table[Date]), -7, DAY)
)

// Moving Average 30-Day
MA30 = 
CALCULATE(
    [Average Price],
    DATESINPERIOD(Date_Table[Date], LASTDATE(Date_Table[Date]), -30, DAY)
)
```

---

#### Comparative Measures

```DAX
// Price vs Market Average
Price vs Market Avg = 
VAR MarketAvg = CALCULATE([Average Price], ALL(Price_Data[Market]))
VAR CommodityPrice = [Average Price]
RETURN CommodityPrice - MarketAvg

// Price Index (Base = 100 in January)
Price Index = 
VAR BasePrice = 
    CALCULATE(
        [Average Price],
        Date_Table[Month] = 1
    )
RETURN DIVIDE([Average Price], BasePrice, 0) * 100

// Rank by Price
Price Rank = 
RANKX(
    ALL(Price_Data[Commodity]),
    [Average Price],
    ,
    DESC,
    DENSE
)

// Top N Commodities
Is Top 10 Commodity = IF([Price Rank] <= 10, TRUE(), FALSE())
```

---

#### Statistical Measures

```DAX
// Standard Deviation
Price Std Dev = STDEV.P(Price_Data[Modal_Price])

// Coefficient of Variation (CV)
Price CV = DIVIDE([Price Std Dev], [Average Price], 0) * 100

// Price Percentiles
Price 25th Percentile = PERCENTILE.INC(Price_Data[Modal_Price], 0.25)
Price 50th Percentile = PERCENTILE.INC(Price_Data[Modal_Price], 0.50)
Price 75th Percentile = PERCENTILE.INC(Price_Data[Modal_Price], 0.75)

// Interquartile Range
IQR = [Price 75th Percentile] - [Price 25th Percentile]
```

---

#### Weather-Related Measures

```DAX
// Average Temperature (if weather data loaded)
Avg Temperature = 
AVERAGE(Price_Data[temperature_mean])

// Total Rainfall
Total Rainfall = SUM(Price_Data[rain_sum])

// Rainy Days Count
Rainy Days = 
CALCULATE(
    COUNT(Price_Data[Arrival_Date]),
    Price_Data[rain_sum] > 0
)

// Temperature-Price Correlation (simplified)
Temp Price Correlation = 
// Note: Requires Python/R visual for true correlation
// This is a directional indicator
VAR TempTable = 
    SUMMARIZE(
        Price_Data,
        Price_Data[Arrival_Date],
        "AvgTemp", AVERAGE(Price_Data[temperature_mean]),
        "AvgPrice", AVERAGE(Price_Data[Modal_Price])
    )
RETURN
    // Return count for visualization
    COUNTROWS(TempTable)
```

---

#### Conditional Formatting Measures

```DAX
// Price Status Color
Price Status Color = 
SWITCH(
    TRUE(),
    [Average Price] < 2000, "Red",
    [Average Price] < 5000, "Yellow",
    [Average Price] < 10000, "Green",
    "Gold"
)

// Volatility Status
Volatility Status = 
SWITCH(
    TRUE(),
    [Average Volatility] < 50, "Stable",
    [Average Volatility] < 100, "Moderate",
    [Average Volatility] < 150, "High",
    "Very High"
)

// Performance Indicator
Performance Indicator = 
IF([MoM % Change] > 0, "‚ñ≤", IF([MoM % Change] < 0, "‚ñº", "‚óè"))
```

---

#### KPI Measures

```DAX
// Price Growth Rate (YoY if multi-year, otherwise MoM)
Growth Rate = [MoM % Change]

// Market Share %
Market Share % = 
DIVIDE(
    CALCULATE([Total Records], ALLSELECTED(Price_Data[Commodity])),
    CALCULATE([Total Records], ALL(Price_Data[Commodity]))
) * 100

// Record Density (records per day)
Record Density = 
DIVIDE(
    [Total Records],
    DISTINCTCOUNT(Price_Data[Arrival_Date]),
    0
)

// Active Commodities (with records in period)
Active Commodities = 
CALCULATE(
    DISTINCTCOUNT(Price_Data[Commodity]),
    Price_Data[Modal_Price] <> BLANK()
)
```

#### Advanced Analytics Measures

```DAX
// Seasonality Index
Seasonality Index = 
VAR AnnualAvg = CALCULATE([Average Price], ALL(Date_Table[Month]))
VAR MonthlyAvg = [Average Price]
RETURN DIVIDE(MonthlyAvg, AnnualAvg, 1) * 100

// Volatility Score (normalized 0-100)
Volatility Score = 
VAR MaxVol = CALCULATE(MAX(Price_Data[Price_Volatility_%]), ALL())
VAR CurrentVol = [Average Volatility]
RETURN DIVIDE(CurrentVol, MaxVol, 0) * 100

// Trend Direction
Trend Direction = 
VAR CurrentPeriod = [Average Price]
VAR PreviousPeriod = [Previous Month Price]
RETURN
    SWITCH(
        TRUE(),
        CurrentPeriod > PreviousPeriod * 1.05, "Strong Up",
        CurrentPeriod > PreviousPeriod, "Up",
        CurrentPeriod < PreviousPeriod * 0.95, "Strong Down",
        CurrentPeriod < PreviousPeriod, "Down",
        "Stable"
    )
```

### Phase 3: Create Calculated Measures (Day 2)

#### ‚è±Ô∏è Estimated Time: 2-3 hours

**Step 9: Create Measure Table**
```
Modeling tab ‚Üí New Table

Type: _Measures = ROW("X", 1)

This creates a blank table to organize all measures.
```

**Step 10: Add Core Measures**

Copy and paste each measure into _Measures table:

```dax
// 1. Total Records
Total Records = COUNTROWS(Price_Data)

// 2. Average Modal Price
Avg Modal Price = AVERAGE(Price_Data[Modal_Price])

// 3. Price Range
Price Range = MAX(Price_Data[Max_Price]) - MIN(Price_Data[Min_Price])

// 4. YoY Growth
YoY Price Growth = 
VAR CurrentYear = MAX(Price_Data[Year])
VAR PrevYear = CurrentYear - 1
VAR CurrentPrice = CALCULATE([Avg Modal Price], Price_Data[Year] = CurrentYear)
VAR PrevPrice = CALCULATE([Avg Modal Price], Price_Data[Year] = PrevYear)
RETURN
DIVIDE(CurrentPrice - PrevPrice, PrevPrice, 0)

// 5. Market Share
Market Share % = 
DIVIDE(
    COUNTROWS(Price_Data),
    CALCULATE(COUNTROWS(Price_Data), ALL(Price_Data[Market]))
) * 100

// 6. Price Volatility Avg
Avg Volatility = AVERAGE(Price_Data[Price_Volatility_%])

// 7. Top Commodity
Top Commodity = 
CALCULATE(
    MAX(Price_Data[Commodity]),
    TOPN(1, ALL(Price_Data), [Avg Modal Price], DESC)
)

// 8. Weather Impact Score
Weather Impact = 
AVERAGE(Weather_Data[temperature_2m_mean]) * 0.4 +
AVERAGE(Weather_Data[precipitation_sum]) * 0.3 +
AVERAGE(Weather_Data[rain_sum]) * 0.3
```

**Step 11: Add Comparison Measures**

```dax
// Month-over-Month
MoM Change = 
VAR CurrentMonth = MAX(Price_Data[Month])
VAR PrevMonth = CurrentMonth - 1
VAR CurrentPrice = CALCULATE([Avg Modal Price], Price_Data[Month] = CurrentMonth)
VAR PrevPrice = CALCULATE([Avg Modal Price], Price_Data[Month] = PrevMonth)
RETURN
DIVIDE(CurrentPrice - PrevPrice, PrevPrice, 0)

// MoM Change % Formatted
MoM % = 
FORMAT([MoM Change], "0.00%")

// Previous Month Price
Prev Month Price = 
CALCULATE(
    [Avg Modal Price],
    DATEADD(DateTable[Date], -1, MONTH)
)
```
